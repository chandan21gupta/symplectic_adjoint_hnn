```python
import numpy as np
import matplotlib.pyplot as plt
import math
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
from torch.autograd import grad
from tqdm import tqdm
torch.manual_seed(42)
import random
from abc import ABC, abstractmethod
random.seed(0)
np.random.seed(0)
```

```python
class Solver(ABC, nn.Module):
    def __init__(self):
        super(Solver, self).__init__()

    @abstractmethod
    def step(self, *args, **kwargs):
        pass

class EulerSolver(Solver):
    def __init__(self, dynamics_fn):
        super(EulerSolver, self).__init__()
        self.dynamics_fn = dynamics_fn
    
    def step(self, p0, q0, dt):
        dp_dt, dq_dt = self.dynamics_fn(p0, q0)
        p = p0 + dt * dp_dt
        q = q0 + dt * dq_dt
        return p, q
```

```python
class Euler:
    
    def __init__(self, dynamics_fn):
        self.dynamics_fn = dynamics_fn
    
    def step(self, p0, q0, dt):
        dp_dt, dq_dt = self.dynamics_fn(p0, q0)
        p = p0 + dt * dp_dt
        q = q0 + dt * dq_dt
        return p, q
    
class LeapFrog:
    
    def __init__(self, dynamics_fn):
        self.dynamics_fn = dynamics_fn
        
    def step(self, p0, q0, dt):
        p_half = p0 + 0.5 * dt * self.dynamics_fn(p0, q0)[0]
        q = q0 + dt * self.dynamics_fn(p_half, q0)[1]
        p = p_half + 0.5 * dt * self.dynamics_fn(p_half, q)[0]
        return p, q
    
class RK2:
    
    def __init__(self, dynamics_fn):
        self.dynamics_fn = dynamics_fn
        
    def step(self, p0, q0, dt):
        p1, q1 = self.dynamics_fn(p0, q0)
        
        # Step 2: Compute k2 for momentum and position
        p2, q2 = self.dynamics_fn(p0 + 0.5 * dt * p1, q0 + 0.5 * dt * q1)
        
        # Step 3: Update momentum and position using weighted average of k1 and k2
        p = p0 + dt * (p1 + p2) / 2 
        q = q0 + dt * (q1 + q2) / 2
        return p, q
    
class SV:
    
    def __init__(self, dynamics_fn):
        self.dynamics_fn = dynamics_fn
        
    def step(self, p0, q0, dt, iterations = 10, p_init = None, q_init = None):
        if p_init:
            p_half = (p_init + p0)/2
        else:
            p_half = p0
        for _ in range(iterations):
            p_half = p0 + 0.5 * dt * self.dynamics_fn(p_half, q0)[0]
        q1 = q0 + 0.5 * dt * self.dynamics_fn(p_half, q0)[1]
        if q_init:
            q2 = (q_init + q1)/2
        else:
            q2 = q1
        for _ in range(iterations):
            q2 = q1 + 0.5 * dt * self.dynamics_fn(p_half, q2)[1]
        p1 = p_half + 0.5 * dt * self.dynamics_fn(p_half, q2)[0]
        # print("sv", p1)
        return p1, q2
    
class PC:
    
    def __init__(self, dynamics_fn):
        self.dynamics_fn = dynamics_fn
        self.explicit = RK2(self.dynamics_fn)
        self.implicit = SV(self.dynamics_fn)
        
    def step(self, p0, q0, dt, iterations):
        p, q = self.explicit.step(p0, q0, dt)
        p, q = self.implicit.step(p0 = p0, q0 = q0, dt = dt, iterations = iterations, p_init = p, q_init = q)
        # print("pc", p)
        return p, q
```

```python
class MassSpring:
    
    def __init__(self, p_min, p_max, q_min, q_max):
        self.pmin = p_min
        self.qmin = q_min
        self.pmax = p_max
        self.qmax = q_max
    
    def h(self, p, q):
        return p**2 + q**2
    
    def dynamics(self, p, q):
        return -2*q, 2*p
    
    def generate_samples(self, num, solver, T, dt, **kwargs):
        dataset = np.zeros((num, T, 3))
        for i in range(num):
            p0 = np.random.rand()*(self.pmax-self.pmin) + self.pmin
            q0 = np.random.rand()*(self.qmax-self.qmin) + self.qmin
            # y0 = np.random.rand(2)*2-1
            # radius = np.random.rand()*0.9 + 0.1
            # y0 = y0 / np.sqrt((y0**2).sum()) * radius
            # p0 = y0[0]
            # q0 = y0[1]
            # q0 = 8.*(np.random.rand()-0.5)
            # p0 = 8.*(np.random.rand()-0.5)
            dataset[i, 0, 0] = p0
            dataset[i, 0, 1] = q0
            dataset[i, 0, 2] = self.h(p0, q0)
            p = p0
            q = q0
            for j in range(1, T):
                p, q = solver.step(p, q, dt, **kwargs)
                dataset[i, j, 0] = p
                dataset[i, j, 1] = q
                dataset[i, j, 2] = self.h(p, q)
        return dataset
    
class Pendulum:
    
    def __init__(self, p_min, p_max, q_min, q_max):
        self.pmin = p_min
        self.qmin = q_min
        self.pmax = p_max
        self.qmax = q_max
    
    def h(self, p, q):
        return p**2 + 3*(1-np.cos(q))
    
    def dynamics(self, p, q):
        return -3*np.sin(q), 2*p
    
    def generate_samples(self, num, solver, T, dt, **kwargs):
        dataset = np.zeros((num, T, 3))
        for i in range(num):
            # y0 = np.random.rand(2)*2-1
            # radius = np.random.rand()*0.9 + 0.1
            # y0 = y0 / np.sqrt((y0**2).sum()) * radius
            # p0 = y0[0]
            # q0 = y0[1]
            p0 = np.random.rand()*(self.pmax-self.pmin) + self.pmin
            q0 = np.random.rand()*(self.qmax-self.qmin) + self.qmin
            dataset[i, 0, 0] = p0
            dataset[i, 0, 1] = q0
            dataset[i, 0, 2] = self.h(p0, q0)
            p = p0
            q = q0
            for j in range(1, T):
                p, q = solver.step(p, q, dt, **kwargs)
                dataset[i, j, 0] = p
                dataset[i, j, 1] = q
                dataset[i, j, 2] = self.h(p, q)
        return dataset

class Tao:
    
    def __init__(self, p_min, p_max, q_min, q_max):
        self.pmin = p_min
        self.qmin = q_min
        self.pmax = p_max
        self.qmax = q_max
    
    def h(self, p, q):
        return 0.5 * (p**2 + 1) * (q**2 + 1)
    
    def dynamics(self, p, q):
        return -q*(p**2 + 1), p*(q**2 + 1)
    
    def generate_samples(self, num, solver, T, dt, **kwargs):
        dataset = np.zeros((num, T, 3))
        for i in range(num):
            p0 = np.random.rand()*(self.pmax-self.pmin) + self.pmin
            q0 = np.random.rand()*(self.qmax-self.qmin) + self.qmin
            # y0 = np.random.rand(2)*3-1
            # radius = np.random.rand()*0.9 + 0.1
            # y0 = y0 / np.sqrt((y0**2).sum()) * radius
            # p0 = y0[0]
            # q0 = y0[1]
            dataset[i, 0, 0] = p0
            dataset[i, 0, 1] = q0
            dataset[i, 0, 2] = self.h(p0, q0)
            p = p0
            q = q0
            for j in range(1, T):
                p, q = solver.step(p, q, dt, **kwargs)
                dataset[i, j, 0] = p
                dataset[i, j, 1] = q
                dataset[i, j, 2] = self.h(p, q)
        return dataset
```
